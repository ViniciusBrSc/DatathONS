## Dados de temperatura

```python
df = pd.read_csv('../../data/raw/inmet.csv', sep=';')
df
```

```python
df['id_varmeteo'].unique()
# Há 3 séries de temperatura disponíveis: temperaturas instantâneas, máximas e mínimas.
# Por constarem como linhas nos registros, devem ser transformadas em colunas. 
```

```python
df[df['din_medicao']=='2020-05-23 09:00:00.000']['id_estacaometeo'].value_counts().to_frame().T
# Para cada hora só há 3 registros de cada estação, um para cada uma das 3 séries de temperatura.
```

```python
df['datetime'] = df['din_medicao'].map(lambda x: datetime.fromisoformat(x))
df
# A coluna de data e hora com formato string deve ser transformada para o formato adequado.
# Assim poderemos ordenar o dataframe.
```

```python
df_ins = df[df['id_varmeteo']=='TEM_INS'].sort_values(['datetime','id_estacaometeo'], ascending=True)
df_min = df[df['id_varmeteo']=='TEM_MIN'].sort_values(['datetime','id_estacaometeo'], ascending=True)
df_max = df[df['id_varmeteo']=='TEM_MAX'].sort_values(['datetime','id_estacaometeo'], ascending=True)
df_ins = df_ins.reset_index().drop('index', axis=1)
df_min = df_min.reset_index().drop('index', axis=1)
df_max = df_max.reset_index().drop('index', axis=1)

# O dataframe é separado em 3, um para cada série de temperatura, e então 
# é ordenando primeiro no tempo e depois segundo o código da estação.

df_max.shape, df_ins.shape, df_min.shape
# Os 3 possuem o mesmo número de linhas.
```

```python
pd.concat([
    df_max[['datetime', 'id_estacaometeo']], 
    df_ins[['datetime', 'id_estacaometeo']], 
    df_min[['datetime', 'id_estacaometeo']]
], axis=1, ignore_index=True)
# Verificando que a concatenação lateral simples une os valores das três
# séries de temperatura corretamente graças ao ordenamento adotado.
```

```python
df_new = df_ins
df_new['temp_ins'] = df_ins['val_medicao']
df_new['temp_min'] = df_min['val_medicao']
df_new['temp_max'] = df_max['val_medicao']
df_new.drop([
    'id_varmeteo', 
    'nom_varmeteo', 
    'din_medicao', 
    'din_inclusaodl', 
    'val_medicao',
    'val_altitude'
], axis=1, inplace=True)
df_new
# Limpeza do dataframe
```

```python
df_new
# df_new.to_csv('../../data/processed/data_nulls.csv')
```

```python
df_new[df_new['temp_ins'].isnull()].shape
```

```python
df_new.isnull().sum().sum(), df_new[['temp_ins','temp_min','temp_max']].isnull().sum().sum()
# Há nulos apenas nas séries de temperatura.
```

```python
df_new['ano'] = df_new['datetime'].dt.year
df_new['mes'] = df_new['datetime'].dt.month
df_new['dia'] = df_new['datetime'].dt.day
df_new['hora'] = df_new['datetime'].dt.hour
df_new
```

### Preenchendo nulos

O método escolhido para preencher os dados faltantes foi o agrupamento pela média de cada estação em cada mês:

```python
df_group = df_new.groupby(['id_estacaometeo', 'mes'])[['temp_ins', 'temp_min', 'temp_max']].mean().reset_index()
df_group
```

```python
df_group.isnull().sum()
```

```python
df_group[df_group['temp_ins'].isnull()]
# Estação A745 ficou um ano inativa
```

```python
df_new[df_new['id_estacaometeo']=='A745']
```

Com isso, unem-se o dataframe e o agrupamento com um *left-join* sobre a chave composta:

```python
df_new_2 = df_new.merge(df_group, how='left', on=['id_estacaometeo', 'mes'])
```

```python
df_new_2['temp_ins_x'].fillna(df_new_2['temp_ins_y'], inplace=True)
df_new_2['temp_min_x'].fillna(df_new_2['temp_min_y'], inplace=True)
df_new_2['temp_max_x'].fillna(df_new_2['temp_max_y'], inplace=True)
df_new_2.drop(['ano', 'mes', 'dia', 'hora'], axis=1, inplace=True)
df_new_2.rename(columns={
    'id_estacaometeo':'estacao',
    'nom_longo':'cidade',
    'val_latitude':'lat',
    'val_longitude':'lon',
    'temp_ins_x':'temp_ins',
    'temp_min_x':'temp_min',
    'temp_max_x':'temp_max',
    'temp_ins_y':'temp_ins_group',
    'temp_min_y':'temp_min_group',
    'temp_max_y':'temp_max_group'
}, inplace=True)
```
